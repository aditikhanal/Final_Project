```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python

base_path = r""

# File paths
admissions_path = rf'{base_path}\hosp\admissions.csv.gz'
patients_path   = rf'{base_path}\hosp\patients.csv.gz'
icustays_path   = rf'{base_path}\icu\icustays.csv.gz'
```

```python
# --- Step 1: Load data with datetime parsing ---
print("Loading data...")
icustays = pd.read_csv(icustays_path, parse_dates=['intime', 'outtime'])
admissions = pd.read_csv(admissions_path, parse_dates=['admittime', 'dischtime', 'deathtime'])
patients = pd.read_csv(patients_path)
```

```python
# --- Step 2: Merge all tables ---
print("Merging data...")
df = icustays.merge(admissions, on='hadm_id', how='left', suffixes=('', '_adm'))
df = df.merge(patients, on='subject_id', how='left')
```

```python
# --- Step 3: Calculate ICU LOS in hours ---
df['los_hours'] = (df['outtime'] - df['intime']).dt.total_seconds() / 3600
```

```python
# --- Step 4: Calculate Age ---
df['anchor_year'] = pd.to_numeric(df['anchor_year'], errors='coerce')
df['admit_age'] = df['anchor_age'] + (df['intime'].dt.year - df['anchor_year'])

# Filter adult patients
df = df[(df['admit_age'] >= 18) & (df['admit_age'] <= 120)]
```

```python
# --- EDA ---

# 1. Summary statistics
print("\n ICU LOS Summary Statistics (hours):")
print(df['los_hours'].describe())
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull(), cbar=False, cmap=sns.color_palette(["white", "red"]))
plt.title("Missing Data Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10, 6)) 
sns.histplot(df['los_hours'], bins=100, kde=True) 
plt.title('Distribution of ICU Length of Stay (Hours)') 
plt.xlabel('LOS (hours)') 
plt.ylabel('Patient Count') 
plt.xlim(0, 500) 
plt.grid(True) 
plt.tight_layout() 
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(np.log1p(df['los_hours']), bins=50, kde=True, color='orange')
plt.title('Log-Transformed ICU LOS Distribution')
plt.xlabel('Log(LOS hours + 1)')
plt.ylabel('Patient Count')
plt.show()
```

```python
# 5. Scatter: LOS vs. Age
sns.boxplot(data=df, x='gender', y='los_hours', palette='Set2')
plt.title('ICU LOS by Gender')
plt.ylim(0, 200)
plt.show()
```

```python
sns.boxplot(data=df, x='age_group', y='los_hours', palette='Set3')
plt.title('ICU LOS by Age Group')
plt.ylim(0, 300)
plt.show()
```

```python
df['age_group'] = pd.cut(df['admit_age'], bins=[18,30,50,70,120])
sns.boxplot(data=df, x='age_group', y='los_hours', hue='gender')
```

```python
df['admit_year'] = df['intime'].dt.year
sns.boxplot(data=df, x='admit_year', y='los_hours', palette='coolwarm')
plt.title('ICU LOS Trend Over Years')
plt.ylim(0, 300)
plt.show()
```

```python
stay_counts = df['subject_id'].value_counts()
plt.figure(figsize=(8,5))
sns.histplot(stay_counts, bins=20, kde=False, color='green')
plt.title('Number of ICU Stays per Patient')
plt.xlabel('ICU Stays')
plt.ylabel('Patient Count')
plt.show()
```

```python
readmitted = (stay_counts > 1).mean() * 100
print(f"Percentage of patients with multiple ICU stays: {readmitted:.2f}%")
```
